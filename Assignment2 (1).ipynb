```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = pd.read_csv('cancer.csv')
data
```

```python
data.shape
```

```python
data.isna().sum()
```

```python
data = data.drop(["id"],axis= 'columns')
```

```python
data
```

```python
data['diagnosis']=data['diagnosis'].map({'M':1,'B':0})
data
```

```python
data.describe()
```

```python
plt.boxplot(data['radius_mean'])
```

```python
plt.boxplot(data['texture_mean'])
```

```python
X = data.iloc[:,1:]
X
```

```python
y = data.iloc[:,0]
y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25)
```

```python
X_train
```

```python
y_train
```

```python
X_test
```

```python
y_test
```

```python
#Using tree classifier for model Building 
from sklearn import tree
tree_clas = tree.DecisionTreeClassifier(criterion='entropy')
tree_clas.fit(X_train,y_train)
```

```python
y_test_predict = tree_clas.predict(X_test)
y_test_predict
```

```python
np.array(y_test)
```

```python
y_train_predict = tree_clas.predict(X_train)
y_train_predict
```

```python
np.array(y_train)
```

```python
from sklearn.metrics import accuracy_score
test_accuracy = accuracy_score(y_test,y_test_predict)
print(test_accuracy)
```

```python
train_accuracy = accuracy_score(y_train,y_train_predict)
print(train_accuracy)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test,y_test_predict)
```
